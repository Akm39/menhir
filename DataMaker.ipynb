```python
import pandas as pd
import glob
import datetime as dt
from tqdm.notebook import tqdm
import warnings
import pickle as pkl

warnings.filterwarnings('ignore')
```

```python
df_vol=pd.read_pickle('Data Storage/volumes.pkl')
```

```python
n=10 #Number of days
```

```python
df_rolling_avg = df_vol.rolling(window=n).mean()
```

```python
dates_all=list(df_rolling_avg.index)
```

```python
files = glob.glob('5_min_bar_data/*')
```

```python
vol_dict={}
open_dict={}
close_dict={}

prog=tqdm(total=len(files))

for file in files:
    data=pd.read_csv(file)
    data['Datetime']=(data['Date']+' '+data['Time'])
    data['DT']=data['Datetime'].apply(lambda x:dt.datetime(int(x[0:4]),int(x[5:7]),int(x[8:10]),int(x[11:13]),int(x[14:16])))
    data.drop(columns=['Time','High','Low','Market Cap','Market Cap Rank','Datetime'],inplace=True)
    for date in data['Date'].unique():
        df=data[data['Date']==date]
        df.drop(columns='Date',inplace=True)
        df.set_index('DT', inplace=True)
        vol_df=df.pivot_table(values='Volume',columns='Ticker',index=df.index)
        open_df=df.pivot_table(values='Open',columns='Ticker',index=df.index)
        close_df=df.pivot_table(values='Close',columns='Ticker',index=df.index)
        vol_dict[date]=vol_df
        open_dict[date]=open_df
        close_dict[date]=close_df
        
    prog.update(1)
```

```python
rv_dict={}

for day in vol_dict.keys():
    index=dates_all.index(day)
    if index==0:
        prev_day=day
    else:
        prev_day=dates_all[index-1]
    rv_dict[day]=((vol_dict[day]/df_rolling_avg.loc[prev_day]).dropna(axis=1, how='all'))
```

```python
with open('Data Storage/vol_dict.pkl','wb') as f:
    pkl.dump(vol_dict,f)
```

```python
with open('Data Storage/close_dict.pkl','wb') as f:
    pkl.dump(close_dict,f)
```

```python
with open('Data Storage/open_dict.pkl','wb') as f:
    pkl.dump(open_dict,f)
```

```python
with open('Data Storage/rv_dict.pkl','wb') as f:
    pkl.dump(rv_dict,f)
```
